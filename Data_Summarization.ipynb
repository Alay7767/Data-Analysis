```python
# Problem 3: Full Data Summarization

nonNumeric = df [['App','Category','Size','Installs','Type','Content Rating','Genres','Last Updated','Current Ver','Android Ver']]
for cols in nonNumeric.columns:
    print(f"\nColumn {cols}:")
    value_counts = nonNumeric[cols].value_counts().head(10)
    for values, counts in value_counts.items():
        print(f" {values}: {counts}")
```

```python
numeric = df [['Rating', 'Reviews', 'Price']]
for cols in numeric.columns:
    print(f"\nColumn {cols}:")
    value_counts = numeric[cols].value_counts().head(10)
    for values, counts in value_counts.items():
        print(f" {values}: {counts}")
```
