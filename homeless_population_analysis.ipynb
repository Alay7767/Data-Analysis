```python
# Challenge 1

import pandas as AP
ap = AP.read_csv("C:\\Assignments\\533\\Homeless\\Population-by-state.csv")
ap
```

```python
ap = ap.iloc[1:].copy()
apYears = [2010, 2011, 2012, 2013, 2014, 2015, 2016]
apYearsColumns = [f"respop7{i}" for i in apYears]
ap[apYearsColumns] = ap[apYearsColumns].apply(AP.to_numeric, errors='coerce')
apPopulation = ap[apYearsColumns].sum()
for i in apYears:
 print(f"{i}-{int(apPopulation[f'respop7{i}'])}")
print(f"Years that has the highest population: {apPopulation.idxmax()[-4:]}: {int(apPopulation.max())}")
```

```python
# Challenge 2

ap["MeanOfPopulationCount"] = ap[apYearsColumns].mean(axis=1).round(2)
ap["VarianceOfPopulationCount"] = ap[apYearsColumns].var(axis=1).round(2)
for i, rows in ap.iterrows():
 print(f"{rows["GEO.display-label"]}-{rows["MeanOfPopulationCount"]}-{rows["VarianceOfPopulationCount"]}")
print(f"\n\nLargest mean state : {ap.loc[ap["MeanOfPopulationCount"].idxmax(), "GEO.display-label"]}")
print(f"Largest variance state: {ap.loc[ap["VarianceOfPopulationCount"].idxmax(), "GEO.display-label"]}")
```

```python
# Challenge 3

ap["AvgRateOfChange"] = ap[apYearsColumns].diff(axis=1).mean(axis=1).round(2)
for i, rows in ap.iterrows():
 print(f"{rows["GEO.display-label"]}-{rows["AvgRateOfChange"]}")
```

```python
# Challenge 4

ap1 = AP.read_csv("C:\\Assignments\\533\\Homeless\\2007-2016-Homelessnewss-USA.csv")
ap1
```

```python
ap1["Count"] = ap1["Count"].apply(AP.to_numeric, errors='coerce')
ap1UniqueCocNo = ap1.groupby("CoC Number")["CoC Name"].first().reset_index()
print(f"Total = {ap1UniqueCocNo.shape[0]}")
for i,x in ap1UniqueCocNo.iterrows():
 print(f"{x["CoC Number"]} - {x["CoC Name"]}")
```

```python
# Challenge 5

ap1["Date"] = AP.to_datetime(ap1["Year"], format="%m/%d/%Y", errors="coerce")
ap1["Year"] = ap1["Date"].dt.year
ap1Years = ap1["Year"].unique()
ap1ascYears = sorted(ap1Years)
print("Year", ", Year ".join(map(str, ap1ascYears)))
print(f"Years total = {len(ap1ascYears)}")
```

```python
# Challenge 6

ap1coc500 = ap1[ap1["CoC Number"].str.contains("500")]
ap1TotalHomeless = ap1coc500[ap1coc500["Measures"] == "Total Homeless"].groupby(["Year", "State"])["Count"].sum().reset_index()
ap1UnshelteredHomeless = ap1coc500[ap1coc500["Measures"] == "Unsheltered Homeless"].groupby(["Year", "State"])["Count"].sum().reset_index()
ap1TotalHomeless = ap1TotalHomeless.rename(columns={"Count": "ap1TotalHomeless"})
ap1UnshelteredHomeless = ap1UnshelteredHomeless.rename(columns={"Count": "ap1UnshelteredHomeless"})
ap1StateYearHomelessCount = AP.merge(ap1TotalHomeless, ap1UnshelteredHomeless, on=["Year", "State"], how="left")
for i, rows in ap1StateYearHomelessCount.iterrows():
 print(f"{rows["Year"]}, {rows["State"]}, {rows["ap1TotalHomeless"]}, {rows["ap1UnshelteredHomeless"]}")
```

```python
# Challenge 7

for s in ap1StateYearHomelessCount["State"].unique():
 print(f"{s}:")
 ap1StateData = ap1StateYearHomelessCount[ap1StateYearHomelessCount["State"] == s]
 for i, rows in ap1StateData.iterrows():
     print(f"\t{rows["Year"]}, {rows["ap1TotalHomeless"]}, {rows["ap1UnshelteredHomeless"]}")
```

```python
# Challenge 8

ap1["Count"] = ap1["Count"].apply(AP.to_numeric, errors="coerce")
ap1TotalHomeless = ap1[ap1["Measures"] == "Total Homeless"]["Count"]
print(f"Average : {ap1TotalHomeless.mean().round(2)}")
print(f"Standard Deviation : {ap1TotalHomeless.std():.2f}")
print(f"Maximum : {ap1TotalHomeless.max()}")
print(f"Minimum : {ap1TotalHomeless.min()}")
print(f"Year having max count : {ap1[ap1["Count"] == ap1TotalHomeless.max()]["Year"].iloc[0]}")
print(f"Year having min count: {ap1[ap1["Count"] == ap1TotalHomeless.min()]["Year"].iloc[0]}")
```
